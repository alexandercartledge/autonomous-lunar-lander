```python
# Play the game Manually

from gym.utils.play import play
import gym

# Create env so that render() returns image arrays
env = gym.make("LunarLander-v2", render_mode="rgb_array")

keys_to_action = {
    ():           0,
    (ord("a"),): 1,
    (ord("w"),): 2,
    (ord("d"),): 3,
}

# Now play() will get real frames to display
play(env, keys_to_action=keys_to_action, zoom=2)
env.close()

```
